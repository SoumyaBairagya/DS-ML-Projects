#DATA CLEANING
####General Steps



####Data Cleaning steps
1. Put NA where there is missing values
2. Remove unneccessary columns (as per the description)
3. Format or Binarize the target or unknown variable
4. Check datatypes, handle columns with Object / string type (either convert to number / remove them)
5. Remove columns which have all missing / NA values
6. Remove rows with missing / NA values greater than 25%
7. Fill missing / NA values using central tendencies
8. Remove those columns which have same value in all rows
9. Remove columns which have different values in all rows
10. Normalize the data

#India

```
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 17,20
from sklearn import preprocessing
#Create DataFrame of a table from csv file
#Step 1 - Put NA where there is missing values
ind_df = pd.read_csv("/content/hospital_beds_per_india_v1.csv",na_values=[""])
ind_df
```

```
#Step 2 - Remove unneccessary columns (as per the description)
# Using the drop() function to remove unwanted columns
# country can be accounted for using the variable name. 
# Removing the latitude and longitude as they are provided to help locate the place on map. We won't use it.
ind_df = ind_df.drop(['country','source','source_url','measure'],axis=1)
ind_df.head()
```

```
#Printing the columns after removing unwanted ones
ind_df.columns
```

```
#Step 3 - Format or Binarize the target or unknown variable
# No such variable present
```

```
#Step 4 - Check datatypes, handle columns with Object / string type (either convert to number / remove them) 
#Step 4.1 - Check datatypes and handle objects
ind_df.dtypes
```

```
#Step 4.2
# column "type" has datatype object
# convert it to numbers
ind_type = ind_df.type.unique()
ind_type
```

```
# Assigning 5 to TOTAL in ind_df col type
ind_df.type[ind_df.type == "TOTAL"] = 5
ind_df.head(10)
```

```
#Step 5 - Remove columns which have all missing / NA values
#Step 5.1 
print(ind_df.isnull().all())
```

```
#Step 5.2
# As we see, in ind_df 'county' is holding all null values.
# remove county col from ind_df as it holds all null values.
l=ind_df.columns[ind_df.isnull().all()]
ind_df.drop(l, inplace=True, axis=1)
ind_df.head(10)
```

```
#Step 6 - Remove rows with missing / NA values greater than 25%
print(ind_df.isnull())
print(ind_df.isnull().sum(axis=1))
ind_df[ind_df.isnull().sum(axis=1)>(ind_df.shape[1]/4)]
ind_df = ind_df[ind_df.isnull().sum(axis=1)\
        <=(ind_df.shape[1]/4)]
```

```
#Step 7
# Print colmuns that have minimum one NaN value
ind_df.columns[ind_df.isna().any()]
```

No NaN values remaining

```
#Step 8
# Check those columns where all values are same
(ind_df != ind_df.iloc[0]).any()
```

```
#Step 9
# Remove columns which have same value in all rows
ind_df=ind_df.loc[:, (ind_df != ind_df.iloc[0]).any()]
```

```
#Step 10 - Normalize data
#Only beds and population are to be normalized. States and year will remain the way they are.
ind_beds_population = ind_df.loc[:,['beds', 'population']]
# beds - no of beds per 1000. Add a column for total number of beds for entire population
ind_beds_population['total_beds'] = ind_beds_population.beds*ind_beds_population.population/1000
ind_beds_population
```

```
#Scale this down to range (0,1)
#x = ind_beds_population.values
#min_max_scaler = preprocessing.MinMaxScaler()
#x_scaled = min_max_scaler.fit_transform(x)
#ind_beds_population = pd.DataFrame(x_scaled,columns=ind_beds_population.columns)
#ind_beds_population
```

Min max normalization Formula  
***(element - min) / (max - min)***

```
# update this beds_popu in ind_df
#ind_df.beds = beds_popu.beds
#ind_df.population = beds_popu.population
ind_df['total_beds'] = ind_beds_population.total_beds
```

```
ind_df['bed_per_person'] = ind_df.total_beds/ind_df.population
```
